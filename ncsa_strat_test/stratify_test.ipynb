```python
def do_cross_tabs(some_data, index_name, col_names):
    data_cols = []
    for col_name in col_names:
        data_cols.append(some_data[col_name].values)
    xtab = pd.crosstab(index = some_data[index_name],
                       columns = data_cols,
                       colnames = col_names,
                       margins = True) / some_data.shape[0]
    return(xtab)

def create_strata_var(some_data, strat_col_names):
    string_data=some_data[strat_col_names].astype(str)
    strat_list=['_'.join(x) for x in string_data[strat_col_names].to_numpy()]
    return(strat_list)
```

```python
do_cross_tabs(ncs, "Age_bin", ["birth_order_cat", "peduc_cat"])
```

```python
strata = create_strata_var(ncs, ["Age_bin", "birth_order_cat", "peduc_cat"])
set(strata)
```

```python
X_train, X_test = train_test_split(ncs, test_size = .2, stratify = strata)
```

```python
do_cross_tabs(X_train, "Age_bin", ["birth_order_cat", "peduc_cat"])
```

```python
do_cross_tabs(X_test, "Age_bin", ["birth_order_cat", "peduc_cat"])
```
